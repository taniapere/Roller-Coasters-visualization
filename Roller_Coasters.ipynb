```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# load rankings data here:
steel_winner = pd.read_csv("Golden_Ticket_Award_Winners_Steel.csv")
wood_winner = pd.read_csv("Golden_Ticket_Award_Winners_Wood.csv")
all_winners = pd.concat([steel_winner, wood_winner])
```

```python
def ranking_through_time(name):
    
    x = all_winners["Year of Rank"][all_winners.Name == name]
    y = all_winners["Rank"][all_winners.Name ==name]
    plt.plot(x, y) 
    plt.xlabel('Years')
    plt.ylabel("Ranks")
    plt.title("Changes of Rank through Time")
    ax = plt.subplot()
    ax.invert_yaxis()
    plt.show()
print(ranking_through_time('El Toro'))      
```

```python
# write function to plot rankings over time for 2 roller coasters here:
def two_through_time(name_one, name_two):
    x_one = all_winners["Year of Rank"][all_winners.Name == name_one]
    y_one = all_winners["Rank"][all_winners.Name ==name_one]
    x_two = all_winners["Year of Rank"][all_winners.Name == name_two]
    y_two = all_winners["Rank"][all_winners.Name ==name_two]

    plt.plot(x_one, y_one, color = "red")
    plt.plot(x_two, y_two, color = 'blue')
    plt.xlabel("Years")
    plt.ylabel("Ranks")
    plt.legend([name_one, name_two])
    ax = plt.subplot()
    ax.invert_yaxis()
    plt.show()
print(two_through_time("El Toro", 'Boulder Dash'))
```

```python
# write function to plot top n rankings over time here:

def n_plots(material, n):
    top_n = material[material['Rank']<=n]
    for coaster in set(top_n["Name"]):
        coaster_data = top_n[top_n["Name"] == coaster]
        plt.plot(coaster_data["Year of Rank"], coaster_data["Rank"], label = coaster, marker = 'o')
    plt.legend()
    plt.title("Top " + str(n) + " rollercoasters")
    plt.xlabel("Year of Rank")
    plt.ylabel("Rank")
    ax = plt.subplot()
    ax.invert_yaxis()
    plt.show
    
    
    
print(n_plots(wood_winner, 5))
```

```python
coasters = pd.read_csv("roller_coasters.csv")
```

```python
# write function to plot histogram of column values here:
def hist(column):
    
    plt.hist(coasters[column])
    
    plt.xlabel(column)
    plt.ylabel("Count")
    plt.title("Histogram of roller coaster " + column)
    plt.show()


print(hist("speed"))
```

```python
# write function to plot inversions by coaster at a park here:
def inver_bar(park):
    coaster = coasters[coasters.park == park]
    
    inversions = coaster.num_inversions
    names = coaster.name
    plt.bar(range(len(names)), inversions)
    
    plt.title("Number of Inversions for each Roller Coaster")
    plt.xlabel("Roller Coaster Name")
    plt.ylabel("Number of Inversions")
    ax = plt.subplot()
    ax.set_xticks(range(len(names)))
    ax.set_xticklabels(names, rotation = 90)
    plt.show()

    
print(inver_bar("Parc Asterix"))
```

```python
# write function to plot pie chart of operating status here:
def pie(df):
    opened = len(df[df.status == "status.operating"])
    closed = len(df[df.status == "status.closed.definitely"]) 
    plt.pie([opened, closed], autopct = "%1d%%")
    plt.axis("equal")
    plt.title("Current Running Condition for Roller Coasters")
    plt.legend(["Operating", "Closed Definitely"])
    plt.show()
    
print(pie(coasters))
```

```python
# write function to create scatter plot of any two numeric columns here:
def scatter(column_one, column_two):
    x = coasters[column_one]
    y = coasters[column_two]
    plt.scatter(x,y)
    plt.title("Relationship between " + column_one + " and " + column_two)
    plt.xlabel(column_one)
    plt.ylabel(column_two)
    plt.show()
    
    
print(scatter("height", "speed"))
```
